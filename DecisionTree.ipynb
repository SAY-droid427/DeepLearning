# Decision Trees

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
salaries = pd.read_csv('salaries.csv')
salaries
```

```python
inputs = salaries.drop('salary_more_than_100k', axis = 'columns')
target = salaries['salary_more_than_100k']
```

```python
inputs
```

```python
target
```

```python
from sklearn.preprocessing import LabelEncoder
le_company = LabelEncoder()
le_job = LabelEncoder()
le_degree = LabelEncoder()
```

```python
inputs['company_n'] = le_company.fit_transform(inputs['company'])
inputs['job_n']  = le_job.fit_transform(inputs['job'])
inputs['degree_n'] = le_degree.fit_transform(inputs['degree'])
inputs
```

```python
inputs_n = inputs.drop(['job','degree','company'], axis = 'columns')
inputs_n
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(inputs_n, target)
X_train
```

```python
model.fit(X_train, y_train)
```

```python
X_test
```

```python
model.predict(X_test)
```

```python
model.score(X_test, y_test)
```
