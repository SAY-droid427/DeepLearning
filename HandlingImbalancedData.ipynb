```python
cols_to_be_updated=['Partner','Dependents','PhoneService','MultipleLines','InternetService','OnlineSecurity','OnlineBackup','DeviceProtection','TechSupport','StreamingTV','StreamingMovies','PaperlessBilling','Churn']
for col in cols_to_be_updated:
  df1[col].replace({'Yes':1,'No':0},inplace=True)
```

```python
df2=pd.get_dummies(data=df1,columns=['InternetService','Contract','PaymentMethod'])
df2.head()
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
cols_to_scale=['MonthlyCharges','TotalCharges','tenure']
df2[cols_to_scale]=scaler.fit_transform(df2[cols_to_scale])
```

```python
df2.shape
```

```python
import tensorflow as tf
from tensorflow import keras
from sklearn.metrics import confusion_matrix, classification_report
```

```python
def ANN(X_train, y_train, X_test, y_test, loss, weights):
  model=keras.Sequential([
      keras.layers.Dense(30,input_dim=26,activation='relu'),
      keras.layers.Dense(15,activation='relu'),
      keras.layers.Dense(1,activation='sigmoid')
  ])

  model.compile(
      optimizer='adam',
      loss=loss,
      metrics=['accuracy']
  )

  if weights==-1:
    model.fit(X_train,y_train,epochs=100)
  else:
    model.fit(X_train,y_train,epochs=100,class_weight=weights)

  print(model.evaluate(X_test,y_test))
  y_preds=model.predict(X_test)
  return y_preds
```

```python
X=df2.drop(['Churn'],axis='columns')
y=df2.Churn
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
y_preds=ANN(X_train,y_train,X_test,y_test,'binary_crossentropy',-1)
```

#Undersampling

```python
count_class_0, count_class_1 = df1.Churn.value_counts()

df_class_0=df2[df2['Churn']==0]
df_class_1=df2[df2['Churn']==1]
```

```python
df_class_0.shape
```

```python
df_class_1.shape
```

```python
df_class_0_under = df_class_0.sample(count_class_1)
df_test_under = pd.concat([df_class_0_under,df_class_1],axis=0)
df_test_under.shape
```

```python
X=df_test_under.drop('Churn',axis='columns')
y=df_test_under.Churn
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,stratify=y)
```

```python
y_preds=ANN(X_train,y_train,X_test,y_test,'binary_crossentropy',-1)
```

#Oversampling

```python
count_class_0,count_class_1
```

```python
df_class_1_over = df_class_1.sample(count_class_0,replace=True)
df_test_over=pd.concat([df_class_0,df_class_1_over],axis=0)
df_test_over.shape
```

```python
X=df_test_over.drop(['Churn'],axis='columns')
y=df_test_over.Churn
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
y_preds=ANN(X_train,y_train,X_test,y_test,'binary_crossentropy',-1)
```

#SMOTE

```python
X=df2.drop(['Churn'],axis='columns')
y=df2.Churn
```

```python
from imblearn.over_sampling import SMOTE
smote=SMOTE(sampling_strategy='minority')
X_sm, y_sm = smote.fit_resample(X,y)
y_sm.value_counts()
```

```python
X_train,X_test,y_train,y_test=train_test_split(X_sm,y_sm,test_size=0.2,stratify=y_sm)
```

```python
y_preds=ANN(X_train, y_train, X_test,y_test,'binary_crossentropy',-1)
```
